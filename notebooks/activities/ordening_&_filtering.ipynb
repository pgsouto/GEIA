```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df1 = pd.read_csv('datasets/csv/COVID-19 Coronavirus.csv', sep = ',', encoding='utf-8')
df1
```

## Classificação de dados por coluna
### Ordenação decrescente do total de mortos de cada país

```python
df1.sort_values(by = 'Total Deaths', axis = 0, ascending = False)
#df1[['Country', 'Total Deaths']].sort_values(by = 'Total Deaths', axis = 0, ascending = False)
```

## Criação de subconjunto com base em condição
### Filtragem de países cuja percentagem de morte é superior a 5%

```python
percentage_ref = 5.0
over_pct_ref_df = df1[df1['Death percentage'] > percentage_ref]
over_pct_ref_df
```

## Extra: gráficos de dispersão e de barras 
### Correlação entre população e total de mortos
### Gŕafico de barras: 10 países com maior número de mortos

```python
#sns.scatterplot(data=df1, x='Population', y='Total Deaths')
sns.regplot(data=df1, x='Population', y='Total Deaths', scatter=True, ci=None)
plt.title('Correlação entre população e total de mortes')
plt.xlabel('População')
plt.ylabel('Total de mortes')
plt.show()

top_10_deaths = df1.sort_values(by = 'Total Deaths', ascending= False).head(10)
sns.barplot(x='Total Deaths', y = 'Country', data=top_10_deaths, color='red')
plt.title('10 países com o maior número de mortos')
plt.xlabel('Total de Mortes')
plt.ylabel('País')
plt.show()
plt.show()
```
