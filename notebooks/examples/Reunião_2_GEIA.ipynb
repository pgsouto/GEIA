## 2.5 Ordenando e filtrando

```python
import pandas as pd

d = {'apples': {'A': 3, 'B': 2, 'C': 0, 'D': 1},
     'oranges': {'A': 0, 'B': 3, 'C': 7, 'D': 2},
     'bananas': {'A': 1, 'B': 3, 'C': 5, 'D': 4},
     'avocados': {'A': 9, 'B': 0, 'C': 0, 'D': 1}}

df = pd.DataFrame(d)
print(df)
```

```python
print(df.sort_index(axis = 0, ascending = False))
print(df.sort_index(axis = 0))
```

```python
#print(df.sort_values(by = 'oranges'))
# print(df.sort_values(by = ['avocados', 'apples']))
print(df.sort_values(by = 'oranges', axis = 0, ascending= False))
```

## 2.6 Combinando colunas

```python
d = {'apples': {'A': 3, 'B': 2, 'C': 0, 'D': 1},
     'oranges': {'A': 0, 'B': 3, 'C': 7, 'D': 2},
     'bananas': {'A': 1, 'B': 3, 'C': 5, 'D': 4},
     'avocados': {'A': 9, 'B': 0, 'C': 0, 'D': 1}}

df = pd.DataFrame(d)
print(df)
```

```python
print(df['bananas'] + df['oranges'])
print(df)
```

```python
df['bananas_oranges'] = df['bananas'] + df['oranges']
print(df)
```

```python
df['bananas_oranges'] = round(df['bananas'] / df['oranges'], 2)
print(df)
```
